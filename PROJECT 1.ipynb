```python
import pandas as pd
import numpy as np
```

```python
atp_matches_2017 = pd.read_csv("atp_matches_2017.csv")
atp_matches_2018 = pd.read_csv("atp_matches_2018.csv")
```

```python
atp_matches_2017.head()
```

```python
atp_matches_2017_left = atp_matches_2017[['surface','draw_size','tourney_level','match_num',
                                        'winner_hand','winner_ht','winner_ioc','winner_age','winner_rank','winner_rank_points',
                                        'loser_hand','loser_ht','loser_ioc','loser_age','loser_rank','loser_rank_points']]
atp_matches_2017_left.columns = ['surface','draw_size','tourney_level','match_num',
                                'left_hand','left_ht','left_ioc','left_age','left_rank','left_rank_points',
                                'right_hand','right_ht','right_ioc','right_age','right_rank','right_rank_points']
atp_matches_2017_left['result'] = 1
atp_matches_2017_left.head()
```

```python
atp_matches_2017_right = atp_matches_2017[['surface','draw_size','tourney_level','match_num',
                                          'loser_hand','loser_ht','loser_ioc','loser_age','loser_rank','loser_rank_points',
                                          'winner_hand','winner_ht','winner_ioc','winner_age','winner_rank','winner_rank_points']]
atp_matches_2017_right.columns = ['surface','draw_size','tourney_level','match_num',
                                'left_hand','left_ht','left_ioc','left_age','left_rank','left_rank_points',
                                'right_hand','right_ht','right_ioc','right_age','right_rank','right_rank_points']
atp_matches_2017_right['result'] = 0
atp_matches_2017_right.head()
```

```python
data_prepared = pd.concat([atp_matches_2017_left,atp_matches_2017_right]).reset_index().drop(['index'],axis=1).dropna()
data_prepared.head()
```

```python
data_new = pd.get_dummies(data_prepared.drop(['surface','tourney_level','left_ioc','right_ioc'],axis=1))
data_new.head()
```

```python
data_new['ht'] = data_new['left_ht'] - data_new['right_ht']
data_new['age'] = data_new['left_age'] - data_new['right_age']
data_new['rank'] = data_new['left_rank'] - data_new['right_rank']
data_new['rank_points'] = data_new['left_rank_points'] - data_new['right_rank_points']
data_new['hand_L'] = data_new['left_hand_L'] - data_new['right_hand_L']
data_new['hand_R'] = data_new['left_hand_R'] - data_new['right_hand_R']
data_new2 = data_new[['ht','age','rank','rank_points','hand_L','hand_R','result']].dropna()
data_new2.head()
```

```python
X = pd.get_dummies(data_prepared.drop(['result'],axis=1)) 
y = data_prepared['result']
```

```python
X.columns
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(X_train, y_train)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.coef_
```
